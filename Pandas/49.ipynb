### 49. Si decimos que un usuario sabe un idioma cuando tiene un nivel de babel mayor o igual a 1, para aquellos que editaron una de las versiones actuales del contenido, ¿Cuál es la tasa de revisiones sin comentario que realizan en función de los idiomas que saben? (⭐⭐⭐)

```python
import numpy as np
import pandas as pd
```

```python
def level_conv(x):
    return np.NaN if x == "N" else np.int32(x)

dfBabel = pd.read_csv("../Dataset/languages.csv", usecols = ["babel_user", "babel_level"], converters = {"babel_level": level_conv})
dfBabel.head()
```

```python
dfContents = pd.read_csv("../Dataset/contents.csv", usecols = ["revisor_id", "revisor_comment"])
dfContents.head()
```

```python
dfBabel["sabe_idioma"] = dfBabel["babel_level"] >= 1
dfBabel.head()
```

```python
dfIdiomasPorUsuario = dfBabel.groupby("babel_user").agg({"sabe_idioma": "sum"})
dfIdiomasPorUsuario.columns = ["idiomas"]
dfIdiomasPorUsuario.head()
```

```python
dfContents["empty_comment"] = dfContents["revisor_comment"].isna()
dfContents.head()
```

```python
dfMerged = dfContents.merge(dfIdiomasPorUsuario,
                            how = "inner",
                            left_on = "revisor_id",
                            right_on = "babel_user")
dfMerged.head()
```

```python
dfComentarios = dfMerged.groupby("idiomas").agg({"empty_comment": ["sum", "count"]})
dfComentarios.columns = ["empty", "total"]
dfComentarios.head()
```

```python
dfComentarios["porcentaje"] = dfComentarios["empty"] / dfComentarios["total"]
dfComentarios.reset_index()[["idiomas", "porcentaje"]]
```

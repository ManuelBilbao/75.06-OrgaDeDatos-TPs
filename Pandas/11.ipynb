### 11. Para todos los comentarios de revisión de contenido que tengan más de 20 ocurrencias realice una matriz cuyas columnas sean esos comentarios y de índice los usuarios/ips con valores: True si ese usuario realizó ese comentario, sino False (⭐⭐) 

```python
import pandas as pd
import numpy as np
```

```python
dfContents = pd.read_csv("../Dataset/contents.csv", usecols = ["revisor_username", "revisor_ip", "revisor_comment"])
dfContents.head()
```

```python
# Unifico revisor_username y revisor_ip en una columna
dfContents["revisor_username"].fillna(dfContents["revisor_ip"], inplace = True)
dfContents.head()
```

```python
# Ocurrencias de un comentario
dfContents["ocurrencias"] = dfContents.groupby("revisor_comment")["revisor_username"].transform("count")
dfContents.head()
```

```python
dfComentarios = dfContents[dfContents["ocurrencias"] > 20]
dfComentarios.head()
```

```python
table = dfComentarios.pivot_table(index = "revisor_username", columns = "revisor_comment")>0
table.head()
```

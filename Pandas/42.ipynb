### 42. Si la experiencia de un usuario es la cantidad de logs en los que participó, ¿cuál es la tasa de contenidos cuya última revisión no tiene comentario en función de la experiencia de su revisor? (⭐⭐⭐)

```python
import numpy as np
import pandas as pd
```

```python
dfLogs = pd.read_csv("../Dataset/logs.csv", usecols = ["contributor_username", "contributor_ip"])
dfLogs.head()
```

```python
dfContents = pd.read_csv("../Dataset/contents.csv", usecols = ["revisor_username", "revisor_ip", "revisor_comment"])
dfContents.head()
```

```python
dfLogs["contributor_username"].fillna(dfLogs["contributor_ip"], inplace = True)
dfContents["revisor_username"].fillna(dfContents["revisor_ip"], inplace = True)
```

```python
dfParticipaciones = dfLogs.groupby("contributor_username").agg({"contributor_username": "count"})
dfParticipaciones.columns = ["experiencia"]
dfParticipaciones.head()
```

```python
dfContents["empty_comment"] = dfContents["revisor_comment"].isna()
dfContents.head()
```

```python
dfMerged = dfContents.merge(dfParticipaciones,
                            how = "left",
                            left_on = "revisor_username",
                            right_on = "contributor_username")
dfMerged.head()
```

```python
dfMerged["experiencia"].fillna(0, inplace = True)
```

```python
dfComentarios = dfMerged.groupby("experiencia").agg({"empty_comment": ["sum", "count"]})
dfComentarios.columns = ["empty", "total"]
dfComentarios.head()
```

```python
dfComentarios["porcentaje"] = dfComentarios["empty"] / dfComentarios["total"]
dfComentarios.head()
```

```python
dfComentarios.reset_index()[["experiencia", "porcentaje"]]
```
